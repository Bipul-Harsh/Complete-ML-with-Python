```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
training_accuracy = []
testing_accuracy = []
k_range = range(1,11)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=0)
```

```python
for k_value in k_range:
    knn = KNeighborsClassifier(n_neighbors=k_value)
    knn.fit(X_train, Y_train)
    training_accuracy.append(knn.score(X_train, Y_train))
    testing_accuracy.append(knn.score(X_test, Y_test))
```

```python
plt.plot(k_range, training_accuracy, label='training accuracy')
plt.plot(k_range, testing_accuracy, label='testing accuracy')
plt.ylabel('Accuracy')
plt.xlabel('k Value/n_neighbors')
plt.legend()
```

## Note to Take:
Here for k=7 even though training accuracy drops, but testing accuracy increases, it tells that for lower number of k the model is very complex and could be a case of overfitting, while if we go beyond k=7 accuracy drops that tell the our model getting more simpler and could lead to underfit.

We have to make a balance between overfitting and underfitting, which here k=7 depicts better.
